```python

# Importação da biblioteca pandas para manipulação de DataFrames
import pandas as pd 
import numpy as np
```

```python

# Leitura do arquivo .csv por meio da biblioteca pandas
students_df = pd.read_csv("predict_students_dropout_and_academic_success.csv", sep = ";")
students_df.columns = students_df.columns.str.strip()
```

```python

# Drop da coluna alvo para manipulação do dataframe
y = students_df["Target"];
students_df.drop("Target", axis = 1, errors = "ignore", inplace = True)
```

```python

# Funções para calculo de medidas de tendência central
def calculo_media(students_df, column):
    n_instancias = len(students_df[column])
    media = students_df[column].sum() / n_instancias
    return media
    
def calculo_mediana(students_df, column):
    valor_central_1 = students_df[column].sort_values().iloc[2211]
    valor_central_2 = students_df[column].sort_values().iloc[2212]
    mediana = (valor_central_1 + valor_central_2)/2
    return mediana
    
def calculo_ponto_medio(students_df, column):
    ponto_medio = (students_df[column].max() + students_df[column].min())/2
    return ponto_medio
    
def calculo_moda(students_df, column):
    moda = students_df[column].mode()[0]
    return moda
```

```python

# Funções para cálculo de medidas de dispersão
def calculo_amplitude(students_df, column):
    amplitude = students_df[column].max() - students_df[column].min()
    return amplitude

def calculo_variancia(students_df, column):
    n_instancias = len(students_df)
    media = calculo_media(students_df, column)
    
    soma_quadrados = 0;
    for i in students_df[column]:
        soma_quadrados += (i - media) ** 2

    variancia = soma_quadrados/n_instancias
    return variancia

def calculo_desvio_padrao(students_df, column):
    desvio_padrao = calculo_variancia(students_df, column) ** 0.5
    return desvio_padrao

def calculo_coef_variacao(students_df, column):
    coef_variacao = (calculo_desvio_padrao(students_df, column)/calculo_media(students_df, column)) * 100
    return coef_variacao
```

```python

# Funções para cálculo de medidas de posição relativa
def calculo_score_z(students_df, column):
    x = calculo_media(students_df, column)
    desvio_padrao = calculo_desvio_padrao(students_df, column)
    score_z = (students_df[column]-x)/desvio_padrao
    return score_z

def calculo_quartil(column, q):
    ordenados = sorted(column)
    n = len(ordenados)
    posicao = (q * (n + 1)) / 4
    if posicao.is_integer():
        quartil = ordenados[int(posicao) - 1]
    else:
        inferior = int(np.floor(posicao))
        superior = int(np.ceil(posicao))
        quartil = (ordenados[inferior - 1] + ordenados[superior - 1]) / 2
    return quartil

def calculo_amplitude_interquartil(column):
    q1 = calculo_quartil(column, 1)
    q3 = calculo_quartil(column, 3)
    amplitude_quartil = q3 - q1
    return amplitude_quartil
```

```python
# Funções para cálculo de medidas de associação
def calculo_covariancia(students_df, column1, column2):
    n = len(students_df)
    media1 = calculo_media(students_df, column1)
    media2 = calculo_media(students_df, column2)
    
    soma_produtos = 0
    for i in range(n):
        soma_produtos += (students_df[column1].iloc[i] - media1) * (students_df[column2].iloc[i] - media2)
    covariancia = soma_produtos / n
    return covariancia
    
def calculo_correlacao(student_df, column1, column2):
    covariancia = calculo_covariancia(students_df, column1, column2)
    desvio_padrao1 = calculo_desvio_padrao(students_df, column1)
    desvio_padrao2 = calculo_desvio_padrao(students_df, column2)
    correlacao = covariancia / (desvio_padrao1 * desvio_padrao2)
    return correlacao
    
```

```python

def medidas_tendencia_central(students_df, column):
    media = calculo_media(students_df, column)
    mediana = calculo_mediana(students_df, column) 
    ponto_medio = calculo_ponto_medio(students_df, column)
    moda = calculo_moda(students_df, column)

    return {
        'Media': media,
        'Mediana': mediana,
        'Ponto Medio': ponto_medio,
        'Moda': moda
    }
    
def medidas_dispersao(students_df, column):
    amplitude = calculo_amplitude(students_df, column)
    variancia = calculo_variancia(students_df, column)
    desvio_padrao = calculo_desvio_padrao(students_df, column)
    coef_variacao = calculo_coef_variacao(students_df, column)
    
    return {
        'Amplitude': amplitude,
        'Variancia': variancia,
        'Desvio Padrao': desvio_padrao,
        'Coef. Variacao': coef_variacao
    }
    
def medidas_posicao_relativa(students_df, column):
    score_z = calculo_score_z(students_df, column)
    menor_score_z = score_z.min()
    maior_score_z = score_z.max()
    quartil1 = calculo_quartil(students_df[column], 1)
    quartil2 = calculo_quartil(students_df[column], 2)
    quartil3 = calculo_quartil(students_df[column], 3)
    amplitude_quartil = calculo_amplitude_interquartil(students_df[column])

    return {
        'Menor Score-Z': menor_score_z,
        'Maior Score-Z': maior_score_z,
        'Quartil 1': quartil1,
        'Quartil 2': quartil2,
        'Quartil 3': quartil3,
        'Amplitude Interquartil': amplitude_quartil
    }

def medidas_associacao(students_df, column1, column2):
    covariancia = calculo_covariancia(students_df, column1, column2)
    print(f'Covariância entre {column1} e {column2}: {covariancia:.4f}')
    correlacao = calculo_correlacao(students_df, column1, column2)
    print(f'Correlação entre {column1} e {column2}: {correlacao:.4f}')
```

```python
resultados_colunas = []

for column in colunas_numericas:
    medidas_tendencia_central(students_df, column)
    medidas_dispersao(students_df, column)
    medidas_posicao_relativa(students_df, column)

    resumo_coluna = {'Coluna': column}
    resumo_coluna.update(medidas_tendencia_central(students_df, column))
    resumo_coluna.update(medidas_dispersao(students_df, column))
    resumo_coluna.update(medidas_posicao_relativa(students_df, column))
    resultados_colunas.append(resumo_coluna)

df_resumo_estatistico = pd.DataFrame(resultados_colunas)
df_resumo_estatistico = df_resumo_estatistico.round(4)

df_resumo_estatistico
```
