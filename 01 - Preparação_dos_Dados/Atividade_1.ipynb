<p style="text-align: justify;">
  O código completo apresentado abaixo corresponde à etapa inicial do processo de Mineração de Dados, tendo como foco a leitura e visualização de um dataset, bem como a limpeza e o tratamento de valores nulos e/ou dados inconsistentes ou duvidosos. Essas etapas são fundamentais para garantir maior qualidade e confiabilidade nas análises realizadas nas fases seguintes do processo de mineração de dados.
</p>

<p style="text-align: justify;">
  Essa etapa torna-se necessária devido à grande quantidade de informações coletadas atualmente em diversos setores da sociedade, especialmente por grandes empresas de tecnologia (Big Techs). Entretanto, nem todos os dados coletados são necessariamente úteis ou relevantes para a análise proposta. Assim, o processo de tratamento de dados tem como objetivo filtrar as informações relevantes, além de corrigir inconsistências, tratar valores ausentes ou ignorar dados que não contribuem para a análise pretendida.
</p>

```python

# Importação das bibliotecas Pandas e Numpy para manipulação de DataFrames e manipulação de arrays
import pandas as pd
import numpy as np
```

```python

# Leitura do arquivo .csv por meio da biblioteca pandas e atribuição à variável original_videogames_df para manipulação dos dados
csv_PATH = 'best_selling_video_games.csv'
original_videogames_df = pd.read_csv(csv_PATH)
processed_videogames_df = pd.read_csv(csv_PATH)
```

```python

# Visualização inicial das informações do DataFrame original, tabelas, contagem de células com valores não nulos e tipo de dados nas colunas
print("\n")
original_videogames_df.info()

# Visualização do DataFrame completo, possível pelo volume de dados ser baixo
print("\n")
original_videogames_df
```

<p style="text-align: justify;">
A leitura inicial da base de dados escolhida revela a presença de informações desnecessárias, dados ruidosos e valores ausentes, tornando necessária a utilização de técnicas e ferramentas de tratamento de dados. Conforme afirmam Castro e Ferrari (2016, p. 105): “Assim, as ferramentas para a limpeza de dados atuam no sentido de imputar valores ausentes, suavizar ruídos, identificar valores discrepantes (outliers) e corrigir inconsistências.”
</p>

<p style="text-align: justify;">Observa-se também que, das 10 colunas presentes no conjunto de dados, 9 são tratadas como dados do tipo String. Mesmo contendo valores numéricos, esses dados estão inicialmente representados como texto, sendo necessária sua conversão para tipos numéricos apropriados em etapas posteriores do tratamento de dados. Além disso, as colunas “Ref.” e “Table_Number” não apresentam informações relevantes para a análise proposta e, portanto, podem ser desconsideradas nas etapas posteriores do processo de análise de dados.
</p>

<p style="text-align: justify;">Durante a inspeção inicial do dataset, também foram identificadas linhas problemáticas, contendo dados desordenados ou incompatíveis com a estrutura do restante da base. Além disso, observam-se alguns erros de digitação e a presença de marcações textuais junto a determinados valores, como por exemplo “[a]”, que acompanham nomes ou categorias e não contribuem para a análise.
</p>

<p style="text-align: justify;">Dessa forma, com base na visualização inicial do conjunto de dados, é possível identificar os principais problemas presentes na base e definir possíveis estratégias para o tratamento dessas inconsistências. Contudo, nota-se que o dataset possui uma quantidade relativamente pequena de registros disponíveis. Assim, o descarte de linhas contendo valores nulos tende a ser evitado, de modo a preservar o maior volume possível de dados, exceto no caso das duas colunas anteriormente mencionadas, que não interferem nas demais informações da base.
</p>

```python

# Retirada das colunas sem informações relevantes para a análise de dados com comando .drop
processed_videogames_df.drop(columns = ["Ref.", "Table_Number"], errors = "ignore", inplace = True)

# "Remoção" dos caracteres do tipo: [a], [b], [k]... dos dados no DatFrame com comando .replace para leitura mais limpa
processed_videogames_df.replace(r"\[.*?\]", "", regex=True, inplace=True)

# Tratamento de Valores nulos para a coluna de 'Series', sendo preenchida com o conteúdo de 'Title' para valores nulos
processed_videogames_df["Series"] = processed_videogames_df["Series"].fillna(processed_videogames_df["Title"])
```

<p style="text-align: justify;">
Ao dispensar as colunas "Ref." e "Table_Number", previamente definidas como irrelevantes para a análise, torna-se possível realizar o tratamento das informações que apresentam erros de escrita, anteriormente mencionados. Esse procedimento consiste na substituição dos caracteres adicionais por um caractere vazio, removendo anotações que não fazem parte do valor original do dado.
</p>

<p style="text-align: justify;">
Um exemplo desse processo pode ser observado no caso em que determinados valores aparecem acompanhados de marcações entre colchetes. Assim, valores como:
</p>

<p style="text-align: center;">
<b>"2011[b]" → "2011"</b>
</p>

<p style="text-align: justify;">
Seguindo esse procedimento e ignorando as linhas problemáticas — entendidas como linhas que apresentam dados fora da ordem ou incompatíveis com a estrutura do restante do dataset, além das duas colunas previamente desconsideradas, observa-se que a única coluna restante contendo valores nulos é a coluna "Series".
</p>

<p style="text-align: justify;">
Com base na análise do contexto da base de dados, que contém informações relacionadas a jogos de plataformas digitais, entende-se que o atributo "Series" representa a franquia ou série à qual um determinado jogo pertence. Como se trata de um dado categórico representado em formato textual, torna-se possível preencher os valores ausentes substituindo-os pelo próprio nome do jogo, informação presente na coluna "Title".
</p>

<p style="text-align: justify;">
Dessa forma, conclui-se que o método de tratamento de dados nulos utilizado neste caso corresponde à imputação manual de valores. Outras abordagens poderiam ser consideradas caso os dados ausentes fossem de natureza numérica. Entretanto, por se tratarem de informações públicas e relativamente conhecidas, como ano de lançamento ou quantidade de vendas, esses valores também poderiam ser obtidos por meio de consulta a fontes externas ou estimados utilizando técnicas estatísticas.
</p>

<p style="text-align: justify;">
Por exemplo, valores numéricos como o número de vendas (em milhões) poderiam ser estimados por meio de médias ou por valores próximos dentro do conjunto de dados, considerando ainda a existência de uma coluna de <i>ranking</i> que estabelece uma ordem coerente entre os jogos, na qual a venda de um título não deve ser inferior à venda do título imediatamente seguinte na classificação.
</p>

```python

# Separação das linhas problemáticas (desordenadas) para facilitação na manipulação dos dados 
problem_indexes = [11, 12, 19, 20, 25, 32, 33, 34]

# Localização das linhas problemáticas pelos índices definidos em problem_indexes e retirada do DataFrame principal com .loc e .drop
# As linhas problemáticas foram armazenadas em um DataFrame secundário com uso do pandas que virou uma matriz com uso do numpy
problem_rows_df = processed_videogames_df.loc[problem_indexes]
processed_videogames_df.drop(index = problem_indexes, inplace = True)
problem_matrix = problem_rows_df.to_numpy()
```

<p style="text-align: justify;">
Os dados obtidos encontram-se numerados de 0 a 49, totalizando 50 registros. Esse volume relativamente baixo de dados permite a realização de um tratamento mais direto e simplificado das linhas que apresentam informações desordenadas. Durante a inspeção inicial da base, foram identificadas apenas 8 linhas com esse tipo de problema. Além disso, embora essas linhas estivessem desordenadas em relação ao restante do <i>dataset</i>, observou-se que elas ainda mantinham uma ordem consistente entre si.
</p>

<p style="text-align: justify;">
Esse tipo de abordagem manual poderia tornar-se inviável em bases de dados maiores ou mais complexas, especialmente em situações nas quais não exista qualquer tipo de padrão ou ordenação entre os registros classificados como problemáticos.
</p>

<p style="text-align: justify;">
Dessa forma, identificou-se que a ordem ideal das colunas do conjunto de dados é:
</p>

<p style="text-align: center;">
<b>
"Rank", "Title", "Platform(s)", "Developer(s)", "Publisher(s)", "Releaseyear", "Sales(millions)", "Series"
</b>
</p>

<p style="text-align: justify;">
Entretanto, nas linhas consideradas desordenadas, a estrutura observada foi a seguinte:
</p>

<p style="text-align: center;">
<b>
"Title", "Sales(millions)", "Releaseyear", "Publisher(s)", "Developer(s)", "Series", "Platform(s)"
</b>
</p>

<p style="text-align: justify;">
Nota-se, portanto, a ausência da coluna correspondente à informação de "Rank" nessas linhas.
</p>

<p style="text-align: justify;">
Durante a implementação do código, essas linhas foram inicialmente separadas e armazenadas em uma matriz. Posteriormente, essa matriz foi convertida em um DataFrame independente, permitindo que os dados fossem reorganizados de acordo com a ordem correta das colunas. Após esse rearranjo, os registros foram reintegrados ao DataFrame original.
</p>

<p style="text-align: justify;">
A informação referente ao atributo "Rank" pôde ser reconstruída utilizando o identificador (ID) de cada linha. Considerando que o conjunto de dados possui 50 registros ranqueados e que os índices do DataFrame variam de 0 a 49, o valor do ranking pode ser obtido somando-se 1 ao índice de cada linha. Assim, o ranking de cada registro corresponde diretamente ao valor de seu índice acrescido de uma unidade.
</p>

```python

# Criação de um vetor limpo para guardar os valores das linhas problemáticas ordenados
new_matrix = []

# Loop para ordenação de linhas problemáticas, arrumando rank de acordo com index e ordem dos dados
for i in range(len(problem_matrix)):
    row = problem_matrix[i]
    rank = problem_indexes[i] + 1

    new_row = [rank, row[0], row[7], row[5], row[3], row[2], row[1], row[6]]
    new_matrix.append(new_row)

# Transformação em matriz do numpy e impressão para visualização dos dados ordenados corretamente
new_matrix = np.array(new_matrix)
print("\n")
print(new_matrix)
print("\n")
```

```python

# Definição das colunas do DataFrame e transformação da matriz de dados limpos em DataFrame para junção dos dados com o DataFrame principal
original_columns = ["Rank", "Title", "Platform(s)", "Developer(s)", "Publisher(s)", "Releaseyear", "Sales(millions)", "Series"]
problem_rows_rebuild_df = pd.DataFrame(new_matrix, columns = original_columns)

# Junção de ambos os DataFrames para finalização da limpeza dos dados
processed_videogames_df = pd.concat([processed_videogames_df, problem_rows_rebuild_df], ignore_index = True)
```

```python

# Conversão de colunas numéricas do tipo 'string' para tipo 'int' e 'float' com .astype
processed_videogames_df["Rank"] = processed_videogames_df["Rank"].astype(int)
processed_videogames_df["Releaseyear"] = processed_videogames_df["Releaseyear"].astype(int)
processed_videogames_df["Sales(millions)"] = processed_videogames_df["Sales(millions)"].astype(float)
```

```python

# Ordenação de todo o Dataframe de acordo com a coluna de rank e reset dos índices para leitura limpa 
processed_videogames_df.sort_values(["Rank"], ascending=True, inplace=True)
processed_videogames_df.reset_index(drop=True, inplace=True)
```

```python

# Leitura final das informações do Dataframe, tipos das colunas ajustados, valores nulos tratados, colunas irrelevantes descartadas
print("\n")
processed_videogames_df.info()
print("\n")
```

<p style="text-align: justify;">
A leitura final do DataFrame após a etapa de tratamento de dados revela a ausência de valores nulos, de modo que todas as colunas passam a possuir 50 registros válidos. Além disso, observa-se que as colunas numéricas agora apresentam tipos de dados apropriados, diferentemente da situação inicial, na qual grande parte das informações estava representada apenas como texto.
</p>

<p style="text-align: justify;">
Outro resultado importante do processo de limpeza é que o conjunto de dados passa a conter apenas informações relevantes para a análise proposta. Erros de escrita foram corrigidos, marcações desnecessárias foram removidas e as linhas anteriormente identificadas como desordenadas foram reorganizadas de acordo com a estrutura correta do dataset.
</p>

<p style="text-align: justify;">
Dessa forma, ao final dessa etapa, o conjunto de dados encontra-se estruturado, consistente e adequado para as etapas posteriores do processo de Mineração de Dados, permitindo a realização de análises mais confiáveis e precisas.
</p>
